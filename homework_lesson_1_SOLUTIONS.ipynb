### 1.3 Basic Arithmetic

### 3.2 Understanding the Working Directory

The **working directory** is the folder where R looks for files and saves files. Think of it as R's "current location" on your computer.

**Key functions:**
| Function | What It Does | Example |
|----------|--------------|--------|
| `getwd()` | **Get** working directory - shows current location | `getwd()` |
| `setwd()` | **Set** working directory - changes location | `setwd("/path/to/folder")` |

**Why this matters:**
- If R can't find your file, it might be looking in the wrong folder
- Use `getwd()` to see where R is looking
- Use `setwd()` to change to the correct folder

**üìù Task:** Write the code to check your current working directory:

````r
# ‚úÖ SOLUTION
# Use the getwd() function to display your current working directory

getwd()
````

### Question 1: Variables

What is the purpose of storing values in variables? Why don't we just use the raw numbers or text directly in our code?

**Your Answer:**

‚úÖ **SOLUTION:** Variables serve several important purposes:

1. **Reusability** - Once a value is stored in a variable, you can use it multiple times without retyping it. If you need to use the same value in 10 different places, you only define it once.

2. **Readability** - A variable name like `customer_age` is much more meaningful than just seeing the number `25` in code. It makes the code self-documenting.

3. **Easy Updates** - If you need to change a value, you only change it in one place (where the variable is defined) rather than finding and updating every instance of the raw value.

4. **Dynamic Values** - Variables can store results of calculations or data that changes during program execution.

### Question 2: Data Structures

What is the difference between a vector and a data frame? When would you use each one?

**Your Answer:**

‚úÖ **SOLUTION:**

**Vector:**
- A one-dimensional collection of values
- All values must be the same type (all numbers OR all text)
- Like a single column or list
- Use when you have a simple list of related items (e.g., test scores, names, prices)

**Data Frame:**
- A two-dimensional structure with rows and columns (like a spreadsheet)
- Different columns can have different types
- Each row is an observation, each column is a variable
- Use when you have tabular data with multiple attributes per observation (e.g., customer records with name, age, and purchase amount)
